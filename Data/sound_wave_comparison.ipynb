```python
import librosa
import matplotlib.pyplot as plt

# Load the audio file
file_path = 'Whisper_base/noisy_phone_call6.wav'  # Replace with your file path
audio_data, sample_rate = librosa.load(file_path)

# Plot the waveform
plt.figure(figsize=(12, 4))
plt.plot(audio_data)
plt.title('Waveform of the audio file')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import librosa
import matplotlib.pyplot as plt

# Load the first audio file
file_path_1 = 'Whisper_base/noisy_phone_call6.wav'
audio_data_1, sample_rate_1 = librosa.load(file_path_1)

# Load the second audio file
file_path_2 = 'Voices/Female American (Nova)/10_words_sample_1.wav'
file_path_2 = 'Whisper_base/noisy_phone_call14.wav'
audio_data_2, sample_rate_2 = librosa.load(file_path_2)

# Plot both waveforms for comparison
plt.figure(figsize=(12, 6))

# Plot the first waveform
plt.subplot(2, 1, 1)
plt.plot(audio_data_1, color='b')
plt.title('Waveform of Noisy Phone Call')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')

# Plot the second waveform
plt.subplot(2, 1, 2)
plt.plot(audio_data_2, color='g')
plt.title('Waveform of Female American Voice (Nova)')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')

# Display the plots
plt.tight_layout()
plt.show()
```
